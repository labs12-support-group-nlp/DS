```python
# clean group urls into group_number_list
groups_df = pd.read_csv('SupportGroupNumbers.csv', header=None)
groups_lst = groups_df[0].values.tolist()

def extract_nums (url_list):
    group_number_list = []
    for i in range (0,len(groups_lst)):
        group_number_list.append(re.findall(r'\d+', groups_lst[i]))
    return group_number_list

group_number_list = extract_nums(groups_lst)
group_number_list 
```

```python
def to_int (list):
    group_number_list = []
    for i in range(0,len(list)):
        group_number_list.append(int(list[i][0]))
    return group_number_list

group_number_list = to_int(group_number_list)
group_number_list
```

```python
# Unix time for n periods of equal length
def back_in_time (days, number, start = time.time()):
    time_list = []
    for i in range(0, number):
        time_list.append(round((start - (i * days * 2592000))))
    return time_list

time_list = back_in_time (30, 12)
time_list
```

```python
# creat list of all permutations of before time and group number
url_var = list(itertools.product(time_list, group_number_list))
len (url_var
```

```python
# use url_var list to create list of urls to scrape

def url_list(url_var = url_var, max_posts = 60):
    urls = []
    for i in range(0, len(url_var)):
        urls.append('https://supportgroups.com/ajax/sg_f/group_past_activity?max=' + str(max_posts) + '&before=' + str(url_var[i][0]) + '&id=' + str(url_var[i][1]))
    return urls

url_list()
```

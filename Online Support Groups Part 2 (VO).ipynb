```python
import re
def remove(list): 
    pattern = 'u[0-9]{4}'
    list = [re.sub(pattern, '', i) for i in list] 
    return ' '.join(list)
```

```python
import numpy as np
def clean(text):
    if len(text) <= 22:
        return np.nan
    text = text.split("'")
    for i in range(len(text)):
        if len(text[i]) <= 3:
            text[i] = ''
    for i in range(len(text)):
        if not text[i]:
            continue
        else:
            text = text[i]
            break
    text = text.split()
    if text[0] == 'n':
        del text[0]
    text = remove(text)
    text = ''.join(text)
#     text = re.sub(r'n(?=[A-Z])', ' ', text)
    text = re.sub(r'[^.][.]n', '. ', text)
    return text
```

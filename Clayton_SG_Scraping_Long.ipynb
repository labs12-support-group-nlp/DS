```python
import pickle
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import time
import itertools
```

```python
group_soup = pickle.load(open('group_soup.pkl', 'br'))
```

```python
# create new list of only pages with content and return count

def clean_pages(list = group_soup):
    clean_list =  []
    for i in range(0, len(list)):
        if str(list[i]).count('/') >= 10:
            clean_list.append(list[i])
    return clean_list

gs = str(clean_pages())
```

```python
def url_list (txt):
    pu = re.findall (r'href=\\u0027https:\\/\\/(.*?)\\u0027', str(txt))
    pu = str(pu).replace("\\", "")
    pu = str(pu).replace("'", '')
    pu = str(pu).replace(" ", '')
    pu = str(pu).replace("[", '')
    pu = pu.split(',')
    return pu

def url_all (soup = group_soup):
    urls = []
    for i in range(0, len(soup)):
        urls.append(url_list(soup[i]))
    return urls

url_all()
post_urls = list(itertools.chain(*url_all()))
post_urls = list(set(post_urls))
post_urls = ['https://' + s for s in post_urls]
lnth = len(post_urls)
print(lnth)
post_urls
```

```python
import urllib.error

hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}


def scrape_soup (url):
    
    try:
        req = Request(url, headers=hdr)
        html_page = urlopen(req)
        soup = BeautifulSoup(html_page, "lxml")
            
    except urllib.error.HTTPError:
        soup = 'blank_data'
    
    return soup


def soup2txt (urls, start):
    for i in range(0, len(urls[start:])):
        dl = scrape_soup (urls[start + i])
        file = open('post' + str(start + i) + '.txt', 'w')
        file.write(str(dl))
                   
soup2txt(post_urls,29989)
```
